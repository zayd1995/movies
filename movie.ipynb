```python
df.drop(columns={'Poster','Cast','Description','Review'}, inplace= True)
```

```python
df
```

```python
df.isna().sum()
```

```python
df['Year'].fillna(method = 'ffill', inplace= True)
```

```python
df['Year'].astype('int64')
```

```python
df['Certificate'] = df['Certificate'].fillna(df.groupby('Genre')['Certificate'].transform('first'))
```

```python
df['Certificate'].fillna('PG-13')
```

```python
df.dropna(subset='Certificate', inplace=True)
```

```python
df['Certificate'].isnull().sum()
```

```python
df.isnull().sum()
```

```python
mean = df['Duration (min)'].mean()
df['Duration (min)'].fillna(mean, inplace= True)
```

```python
df['Duration (min)'].isnull().sum()
```

```python
median = df['Rating'].median()
df['Rating'].fillna(median, inplace= True)
```

```python
meta_med = df['Metascore'].median()
df['Metascore'].fillna(meta_med, inplace= True)
```

```python
df.dropna(subset= 'Director', inplace=True)
```

```python
df.dropna(subset= 'Votes', inplace=True)
```

```python
df.dropna(subset= 'Review Count',inplace= True)
```

```python
df.drop(columns='Review Title', inplace= True)
```

```python
df.isnull().sum()
```

```python
df.head(2)
```

```python
df[['Year','Duration (min)','Metascore']] = df[['Year','Duration (min)','Metascore']].astype('int16')
```

```python
df
```

```python
df['Rating'].describe()
```

```python
filtered_df = df[(df['Rating'] == 9.7) | (df['Rating'] == 1.3)]
filtered_df
```

```python
# Remove commas from the 'Votes' column
df['Votes'] = df['Votes'].str.replace(',', '')

# Convert the 'Votes' column to float
df['Votes'] = df['Votes'].astype('int64')
```

```python
df
```

```python
min = df['Votes'].min()
max = df['Votes'].max()
```

```python
votess = df[(df['Votes'] == min) | (df['Votes'] == max)]
votess
```

```python
hour = df['Duration (min)'] // 60
mint = df['Duration (min)'] % 60
hourmin = hour.astype(str) + 'H ' + mint.astype(str) + 'm'
df['Duration (min)'] = hourmin
```

```python
df
```

```python
maxt = df['Duration (min)'].max()
mint = df['Duration (min)'].min()
df[(df['Duration (min)'] == maxt) | (df['Duration (min)'] == mint)]
```
